```python
def set_def(var, dim):
  for i in range(dim-1):
    for j in range(dim-1):
      var[i,j] = False
  return var
```

```python
def small_loop(variables, dim):
  for i in range(dim-1):
    for j in range(dim-1):
      if variables[i][j]== 0 and variables[i][j+1]== 1 and variables[i+1][j]== 1 and variables[i+1][j+1]== 0:
        return True
  return False
```

```python
def solve_slant_puzzle(slant_puzzle):
    # Get the dimensions of the puzzle grid
    n_rows, n_cols = len(slant_puzzle), len(slant_puzzle[0])
    num_const = 0 

    # Initialize the CP-SAT solver
    model = cp_model.CpModel()

    # Define the variables
    variables = {}
    for i in range(n_rows-1):
        for j in range(n_cols-1):
            #variables[i, j] = model.NewBoolVar(f'({i},{j})')
            variables[i, j] = model.NewIntVar(0, 1, f'({i},{j})')

    # Define the constraints
    for i in range(n_rows):
        for j in range(n_cols):
            if slant_puzzle[i][j] == '*':
                continue

            if i == 0 and j == 0: # Left-Top Corner
                model.Add(variables[i,j] == int(slant_puzzle[i][j]))
                num_const+=1

            elif i == 0 and j == (n_cols-1): # Rigt-Top Corner
                model.Add((1 - variables[i,j-1]) == int(slant_puzzle[i][j]))
                num_const+=1

            elif i == (n_rows-1) and j == 0: # Left-Bottom Corner
                model.Add((1 - variables[i-1,j]) == int(slant_puzzle[i][j]))
                num_const+=1

            elif i == (n_rows-1) and j == (n_cols-1): # Right-Bottom Corner
                model.Add(variables[i-1,j-1] == int(slant_puzzle[i][j]))
                num_const+=1
            
            elif j == 0: # Left Side
                model.Add((1 - variables[i-1,j]) + variables[i,j] == int(slant_puzzle[i][j]))
                num_const+=1

            elif j == (n_cols-1): # Right Side
                model.Add((1 - variables[i,j-1]) + variables[i-1,j-1] == int(slant_puzzle[i][j]))
                num_const+=1

            elif i == 0: # Top Side
                model.Add(variables[(i,j)] + (1 - variables[(i,j-1)]) == int(slant_puzzle[i][j]))
                num_const+=1

            elif i == (n_rows-1): # Bottom Side
                model.Add(variables[i-1,j-1] + (1 - variables[i-1,j]) == int(slant_puzzle[i][j]))
                num_const+=1

            else: # Middle Side
                model.Add(variables[i-1,j-1] + (1 - variables[i-1,j]) + variables[i,j] + (1 - variables[i,j-1]) == int(slant_puzzle[i][j]))
                num_const+=1

    solver = cp_model.CpSolver()
    solution_printer = SlantPuzzleSolutionPrinter(variables)
    solver.parameters.enumerate_all_solutions = True
    solver.Solve(model, solution_printer)

    print(solver.SolutionInfo)

    print('\nStatistics')
    print(f'  conflicts      : {solver.NumConflicts()}')
    print(f'  branches       : {solver.NumBranches()}')
    print(f'  wall time      : {solver.WallTime()} s')
    print(f'  solutions found: {solution_printer.solution_count()}')
    print(f'  number of constraints: {num_const}')
    print(f'  number of variables: {(n_rows-1)*(n_rows-1)}')
```
